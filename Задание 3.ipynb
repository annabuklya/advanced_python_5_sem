```python
import pandas as pd
import os
import csv


files = os.listdir(r"/home/anna/Рабочий стол/advanced python 5 sem/vitamins/")
#print(names) я так и не поняла, почему он дает файлы не в том порядке, в котором они лежат в папке :(

files.sort()
#print(files)

os.chdir(r"/home/anna/Рабочий стол/advanced python 5 sem/vitamins/")

data = [['vitamin', 'vitamers', 'solubility', 'daily_requirement', 'deficiency_diseases']]

for file in files:
    current_file = open(file, 'r')
    data.append(current_file.readlines())
    current_file.close()
    
print(data)
new_data = []

for arr in data:
    new_arr = []
    for i in arr:
        new_arr.append(i.replace("\n", ""))
    new_data.append(new_arr)
    

print(new_data) #это я убрала \n символы из массива, возможно не самый хороший способ



with open('vitamins.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(new_data)
    

df = pd.read_csv('vitamins.csv', sep=",")
df.head(10)
```
